```python
from statistics import correlation

import kagglehub
from seaborn import heatmap
from sympy import false

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_books = pd.read_csv(f'{path}/books.csv')
print(df_books.shape)
df_books.head(3)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()

sn.heatmap(df_books.isna().transpose(), cbar=False, ax=ax)
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()
```

```python
import numpy as np

df_books['missing_description'] = np.where(df_books['description'].isna(), 1, 0)
df_books['age_of_book'] = 2024 - df_books['published_year']
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = df_books[columns_of_interest].corr(method='spearman')

sn.set_theme(style='white')

plt.figure(figsize=(8,6))

heatmap = sn.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'label': 'Spearman correlation'})
heatmap.set_title('Correlation heatmap')
plt.show()
```

```python
books_missing = df_books[~(df_books['description'].isna()) &
                         ~(df_books['num_pages'].isna()) &
                         ~(df_books['average_rating'].isna()) &
                         ~(df_books['num_pages'].isna()) &
                         ~(df_books['published_year'].isna())
                        ]
```

```python
books_missing_25_words = books_missing[books_missing['words_in_description'] >= 25]
books_missing_25_words.shape
```

```python
books_missing_25_words['title_and_subtitle'] = (
    np.where(books_missing_25_words['subtitle'].isna(), books_missing_25_words['title'],
             books_missing_25_words[['title', 'subtitle']].astype(str).agg(': '.join, axis=1))
)
books_missing_25_words
```

```python
books_missing_25_words['tagged_description'] = books_missing_25_words[['isbn13', 'description']].astype(str).agg(" ".join, axis=1)
books_missing_25_words
```

```python
(
    books_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis='columns')
    .to_csv('cleaned_books.csv', index=False)
)
```
