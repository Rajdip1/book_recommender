```python
import numpy as np

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

def calculate_max_emotion_scores(predictions):
    # predictions = list of [ {label, score}, ... ] per sentence
    per_emotion_scores = {label: [] for label in emotion_labels}

    for sent_pred in predictions:          # predictions for ONE sentence
        if not sent_pred:                  # empty prediction list → skip
            continue
        for item in sent_pred:             # each {label, score} dict
            label = item["label"]
            score = item["score"]
            if label in per_emotion_scores:
                per_emotion_scores[label].append(score)

    # take max per emotion; if no scores at all, use 0.0
    return {
        label: (max(scores) if scores else 0.0)
        for label, scores in per_emotion_scores.items()
    }

```

```python
for i in range(10):
    desc = df["description"][i]

    # skip missing / empty descriptions
    if not isinstance(desc, str) or not desc.strip():
        continue

    isbn.append(df["isbn13"][i])

    # split into sentences and drop empty chunks
    sentences = [s for s in desc.split(".") if s.strip()]

    predictions = classifier(sentences)           # list of lists
    max_scores = calculate_max_emotion_scores(predictions)

    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

for i in tqdm(range(len(df))):
    desc = df["description"][i]

    # skip missing / empty descriptions
    if not isinstance(desc, str) or not desc.strip():
        continue

    isbn.append(df["isbn13"][i])

    # split into sentences and drop empty chunks
    sentences = [s for s in desc.split(".") if s.strip()]

    predictions = classifier(sentences)           # list of lists
    max_scores = calculate_max_emotion_scores(predictions)

    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
df_emotions = pd.DataFrame(emotion_scores)
df_emotions['isbn13'] = isbn
```

```python
df_emotions
```

```python
df = pd.merge(df, df_emotions, on="isbn13")
df
```

```python
df.to_csv('books_with_emotions.csv', index=False)
```
